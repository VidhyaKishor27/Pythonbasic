```python
#Merge dataframe using merge()
# how = 'inner' -> give similar items in both x & y i.e x intersection y
# how = 'outer' -> give all items in both x & y i.e x union y
# how = 'left' -> give all items in both x(left dataframe)i.e x completely (including similar items in y)
# how = 'right' -> give all items in both y(right dataframe)i.e y completely (including similar items in x)
import pandas as pd
df1= pd.DataFrame({'ID':[101,102,103,104],'Name':['Raja','Hari','Priya','Kani'],'Edu':['Engineering','Engineering','Arts','Electrical']})
df2=pd.DataFrame({'ID':[101,102,105,106],'Name':['Raja','Hari','Tamil','Sita'],'Exp':['15 yrs','5 yrs','20 yrs','2 yrs']})
print(df1)
print(df2)
df1.merge(df2)
df1.merge(df2, how='inner', on=['ID','Name']) # give same result as for above command.
df1.merge(df2, how='outer',on='ID')
df1.merge(df2, how='left',on=['ID','Name'])
```

```python
#Join dataframe using join()

import pandas as pd
df1= pd.DataFrame({'ID':[101,102,103,104],'Name':['Raja','Hari','Priya','Kani'],'Edu':['Engineering','Engineering','Arts','Electrical']})
df2=pd.DataFrame({'ID':[101,102,105,106],'Name':['Raja','Hari','Tamil','Sita'],'Exp':['15 yrs','5 yrs','20 yrs','2 yrs']})
#print(df1)
#print(df2)
df1.join(df2,how="inner",lsuffix="_left",rsuffix="_right")
df1.join(df2,on='ID',how="outer",lsuffix="_left",rsuffix="_right")
```

```python
#Concatenate dataframe using pd.concate() - adding rows or col in dataframes from another dataframes
import pandas as pd
df1= pd.DataFrame({'ID':[101,102,103,104],'Name':['Raja','Hari','Priya','Kani'],'Edu':['Engineering','Engineering','Arts','Electrical']})
df2=pd.DataFrame({'ID':[101,102,105,106],'Name':['Raja','Hari','Tamil','Sita'],'Exp':['15 yrs','5 yrs','20 yrs','2 yrs']})
#print(df1)
#print(df2)
pd.concat([df1,df2])
pd.concat([df1,df2], join="inner") #gives back only similar coloums as axis is defacult in 0(col)
pd.concat([df1,df2], join="outer") #gives back all items in both dataframes, axis is defacult in 0(col)
pd.concat([df1,df2], join="outer",axis=1) # concatenate both dataframes side by side keeping row index as common if axis =1  (row)
```
