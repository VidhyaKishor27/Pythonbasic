```python
#using matplotlib
import matplotlib.pyplot as plt
import numpy as np
x = np.array([10,20,30,40])
y = np.array ([11,22,33,44])
plt.plot(x,y,'o--g')
plt.show()
```

```python
#matplotlib using file
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/content/drive')
df=pd.read_excel('/content/drive/MyDrive/Python training/Student.xlsx')
print(df)
x=df['Student Name']
y=df['Eng Mark']
zlabel=df['Student Name']
plt.bar(x,y)
plt.plot(df['Student Name'],df['Eng Mark'])
plt.title('Marks')
plt.show()
plt.pie(df['Maths Mark'],labels=zlabel)
plt.show()
```
