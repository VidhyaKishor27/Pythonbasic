```python
#using matplotlib
import matplotlib.pyplot as plt
import numpy as np
x = np.array([10,20,30,40])
y = np.array ([11,22,33,44])
pie_label=["car1","car2","car3","car4"]
pie_explode=[0,0,0,0.1]
plt.plot(x,y,'o--g')
plt.bar(x,y)
plt.xlabel('cars')
plt.ylabel('speed')
plt.title('speed of car')
plt.show()
plt.pie(x,labels=pie_label,explode=pie_explode)
plt.legend(title="Cars")
plt.show()
```
